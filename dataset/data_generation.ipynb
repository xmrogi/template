```python
"""分散データセット

Returns:
   x: 生成されたサンプル (n_samples x n_features)
   y: クラスラベル (n_samples)
   centers: 生成されたクラスの中心座標のリスト
"""
from sklearn.datasets import make_blobs

x, y, centers = make_blobs(n_samples=100,  # サンプル数 
                           n_features=2,  # 特徴量の数
                           centers=3,  # クラスタの数
                           cluster_std=1.5,  # クラスタの標準偏差
                           center_box=(-10., 20.),  # 各クラスタ中心の境界ボックス
                           shuffle=False,  # データセットをシャッフルする
                           return_centers=True,  # 各クラスの中心座標を返す
                           random_state=None,
                           )
```

```python
print(x[:2])
print(y)
print("centers =", centers)
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.scatter(x[:, 0], x[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
"""クラス分類データセット"""
from sklearn.datasets import make_classification

x, y = make_classification(n_samples=1200,  # サンプル数
                           n_features=3,  # 特徴量数
                           n_informative=2,  # 情報価値のある特徴量の数(目的変数のラベルと相関が強い特徴量数)
                           n_redundant=0,  #  冗長な特徴量の数
                           n_repeated=0,  # 複製する特徴量の数
                           n_classes=3,  # クラス数
                           n_clusters_per_class=1,  # クラスごとのクラスタの数
                           weights=None,  # クラスごとのデータの割合
                           flip_y=0.05,  # ランダムに交換されるデータの割合
                           class_sep=1.0,  # ハイパーキューブサイズを乗算する係数
                           hypercube=True,  # クラスタをハイパーキューブの頂点に配置する
                           shift=0.0,  # 指定した値だけ特徴量をシフトする
                           scale=1.0,  # 指定した値で特徴量を乗算する
                           shuffle=False,  # データセットをシャッフルしない
                           random_state=None,
                           )
```

```python
print(x[:3])
print(y[:20])
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.set_title("make_classification : n_features=2  n_classes=3")
ax.scatter(x[:, 0], x[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""ランダム回帰データセット

Returns:
   x: 入力サンプル (n_samples x n_features)
   y: 出力サンプル (n_samples or n_samples x n_targets)
   coef: 線形回帰係数 (n_samples or n_samples x n_targets)
"""
from sklearn.datasets import make_regression

x, y, coef = make_regression(n_samples=100,  # サンプル数
                             n_features=4,  # 特徴量数
                             n_informative=4,  # 情報価値のある特徴量の数(目的変数のラベルと相関が強い特徴量数)
                             n_targets=1,  # 出力次元数
                             noise=10.0,  # ガウシアンノイズの標準偏差
                             bias=0.0,  # バイアス
                             shuffle=True,  # データセットをシャッフルする
                             coef=True,  # 線形モデルの重みを返す
                             random_state=None  # 乱数生成用のシード
                             )
```

```python
print(x[:5])
print(y[:5])
print("coef =", coef)

plt.figure(figsize=(20, 4))
plt.subplot(1, 4, 1)
plt.title("Feature 1")
plt.plot(x[:, 0], y, "bo")
plt.subplot(1, 4, 2)
plt.title("Feature 2")
plt.plot(x[:, 1], y, "ro")
plt.subplot(1, 4, 3)
plt.title("Feature 3")
plt.plot(x[:, 2], y, "go")
plt.subplot(1, 4, 4)
plt.title("Feature 4")
plt.plot(x[:, 3], y, "yo")
```

```python
"""等方性ガウス分布データセット

分位点でラベル付け

Returns:
   x: 生成されたサンプル (n_samples x n_features)
   y: クラスラベル (n_samples)
"""
from sklearn.datasets import make_gaussian_quantiles

x, y = make_gaussian_quantiles(n_samples=500,  # データ件数
                               n_features=2,  # 各サンプルの特徴量数
                               n_classes=3,  # クラス数
                               mean=(0, 0),  # 多次元正規分布の平均(Noneの場合は(0, 0, ...))
                               cov=1,  # 共分散行列の対角成分(共分散は0)
                               shuffle=True,  # データセットをシャッフルする
                               random_state=None,
                               )
```

```python
print(x[:3])
print(y[:10])
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.scatter(x[:, 0], x[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
"""円形データセット(2次元、2クラス)

Returns:
   x: 生成されたサンプル (n_samples x 2)
   y: クラスラベル (n_samples)
"""
from sklearn.datasets import make_circles

x, y = make_circles(n_samples=100,  # データ件数
                    factor=0.5,  # 内側と外側の円の間のスケール係数
                    noise=0.04,  # ガウシアンノイズの標準偏差
                    shuffle=True,  # データセットをシャッフルする
                    random_state=1,  # 乱数生成用のシード
                    )
```

```python
print(x[:2])
print(y)
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.scatter(x[:, 0], x[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
"""半円データセット(2次元、2クラス)

Returns:
   x: 生成されたサンプル (n_samples x 2)
   y: クラスラベル (n_samples)
"""
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=100,  # データ件数
                  noise=0.1,  # ガウシアンノイズの標準偏差
                  shuffle=True,  # データセットをシャッフルする
                  random_state=1,  # 乱数生成用のシード
                  )
```

```python
print(x[:2])
print(y)
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.scatter(x[:, 0], x[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
"""s字カーブデータセット

Returns:
   x: 生成されたサンプル (n_samples x 3)
   color: 多様体の主な次元によるサンプルの単変量座標 (n_samples)
"""
from sklearn.datasets import make_s_curve

x, color = make_s_curve(n_samples=1000,  # データ件数
                        noise=0.01,  # ガウシアンノイズの標準偏差
                        random_state=None,  # 乱数生成用のシード
                        )
```

```python
print(x[:3])
print(color[:5])

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=color, cmap=plt.cm.Spectral)
ax.view_init(10, 60)
```

```python
"""スイスロールデータセット

Returns:
   x: 生成されたサンプル (n_samples x 3)
   color: 多様体の主な次元によるサンプルの単変量座標 (n_samples)
"""
from sklearn.datasets import make_swiss_roll

x, color = make_swiss_roll(n_samples=1000,  # データ件数
                           noise=0.01,  # ガウシアンノイズの標準偏差
                           random_state=None,  # 乱数生成用のシード
                           )
```

```python
print(x[:3])
print(color[:5])

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=color, cmap=plt.cm.Spectral)
ax.view_init(10, 70)
```
